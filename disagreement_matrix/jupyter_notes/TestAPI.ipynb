# read Data

## read scores

```python
def readFile(root, fname):
    with open(root+fname, 'r') as f:
        data =f.readlines()
        data = [x.split() for x in data]
        data = [[-float(x) for x in line] for line in data]
        data = np.asarray(data)
    return data

def readLabel(root, file):
    with open(root + file, 'r') as d:
        label = d.readlines()
        label = np.asarray([1 - int(x[0]) for x in label])
    return label
```

```python
score_fnames = dict()
path = "../../scoresForEnsemble/"
for root, path, files in os.walk(path):
    for file in files:
        if file.endswith(".txt"):
            pureName = file.split("_LOF_score_matrix")[0]
            if pureName not in score_fnames:
                score_fnames[pureName] = dict()
                score_fnames[pureName][file[:-4]] = readFile(root,file)
            else:
                score_fnames[pureName][file[:-4]] = readFile(root,file)
```

## read labels

```python
label_fnames = dict()
path = "../../datasets_resend/"
for root, path, files in os.walk(path):
    for file in files:
        if file.endswith("_label.txt"):
            pureName = file.split("_label")[0]
            label_fnames[pureName] = readLabel(root, file)
```

```python
# get intersection
intersection_dataset = set(label_fnames.keys()).intersection(score_fnames.keys())
len(intersection_dataset)
label_dict = dict()
scores_dict = dict()
for key in intersection_dataset:
    label_dict[key] = label_fnames[key]
    scores_dict[key] = score_fnames[key]
```

```python
res_dict = dict()
auc_dict = dict()
normalizer = RobustScaler()
for key in intersection_dataset:
    y = label_dict[key]
    res_dict[key] = dict()
    auc_dict[key] = dict()
    for fname in scores_dict[key].keys():
        if fname.endswith("LOF_score_matrix"):
            X = scores_dict[key][fname][np.arange(10,21), :].T
        else:
            X = scores_dict[key][fname].T
            
        deem_ensemble = DEEM(X, normalizer, 'auto',label=y, main_detector_index=None)
        res_dict[key][fname] = deem_ensemble.get_ensemble_score()["ensem_sodi:thr_2std:dis_std"]
        auc_dict[key][fname] = metrics.roc_auc_score(y, res_dict[key][fname].values)
```
